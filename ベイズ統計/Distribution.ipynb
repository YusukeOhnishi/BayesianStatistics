```python
#ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#一様分布
from scipy.stats import uniform

X=np.arange(-2,2,0.1)
dist_1=uniform.pdf(X,-1.5,3.0)

plt.plot(X,dist_1,label='a=-1.5,b=1.5')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probabirity')
plt.savefig('./figure/fig_uniform.jpg')
plt.show()
```

```python
#Beta分布
from scipy.stats import beta

X=np.arange(0,1,0.02)
dist_1=beta.pdf(X,7,2)
dist_2=beta.pdf(X,5,5)
dist_3=beta.pdf(X,2,7)

plt.plot(X,dist_1,label='α=7,β=2')
plt.plot(X,dist_2,label='α=5,β=5')
plt.plot(X,dist_3,label='α=2,β=7')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probabirity')
plt.savefig('./figure/fig_beta.jpg')
plt.show()
```

```python
#正規分布
from scipy.stats import norm

X=np.arange(-5,5,0.1)
dist_1=norm.pdf(X,0,1)
dist_2=norm.pdf(X,-2,2)
dist_3=norm.pdf(X,1,1.5)

plt.plot(X,dist_1,label='μ=0,σ=1')
plt.plot(X,dist_2,label='μ=-2,σ=2')
plt.plot(X,dist_3,label='μ=1,σ=1.5')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probabirity')
plt.savefig('./figure/fig_norm.jpg')
plt.show()
```

```python
#指数分布
from scipy.stats import expon

X=np.arange(0,5,0.1)
dist_1=expon.pdf(X,0,1)
dist_2=expon.pdf(X,0,2)
dist_3=expon.pdf(X,0,3)

plt.plot(X,dist_1,label='Θ=1')
plt.plot(X,dist_2,label='Θ=2')
plt.plot(X,dist_3,label='Θ=3')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probabirity')
plt.savefig('./figure/fig_expon.jpg')
plt.show()
```

```python
#ポアソン分布
from scipy.stats import poisson

X=np.arange(0,20,1)
dist_1=poisson.pmf(X,5)

plt.bar(X,dist_1,label='λ=5')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probabirity')
plt.savefig('./figure/fig_poisson.jpg')
plt.show()
```

```python
#Gamma分布
from scipy.stats import gamma

X=np.arange(0,10,0.1)
dist_1=gamma.pdf(X,a=1,scale=2)
dist_2=gamma.pdf(X,a=2,scale=2)
dist_3=gamma.pdf(X,a=3,scale=2)

plt.plot(X,dist_1,label='k=1,Θ=2')
plt.plot(X,dist_2,label='k=2,Θ=2')
plt.plot(X,dist_3,label='k=3,Θ=2')
plt.legend()
plt.xlabel('x')
plt.ylabel('Probabirity')
plt.savefig('./figure/fig_gamma.jpg')
plt.show()
```
