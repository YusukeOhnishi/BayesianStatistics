```python
#ライブラリインポート
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#グラフの見た目変更
plt.style.use("ggplot")

#乱数のSeedを固定
np.random.seed(999)
```

```python
#乱数の数を指定
N_monte=1000

#二次元上の円(中心(0,0)、半径1)の円を考える。
#また計算領域はx,yそれぞれ0~1の間とする。そのため、0～1の間に一様乱数で点を生成する。
x_data=np.random.rand(N_monte)
y_data=np.random.rand(N_monte)

#図の描画
figure,axes=plt.subplots(figsize=(6,6))
draw_circle=plt.Circle((0,0),1)
axes.set_aspect(1)
axes.add_artist(draw_circle)
plt.plot(x_data,y_data,'.')
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
```

```python
#上記の青枠内の領域の面積を求めるため、入った点の数の割合を考える。
#この割合が(青枠領域内の面積)/(正方形1×1の面積)と一致するはずである。

#入った青枠内に入った点の割合計算
sampling_rate=np.sum((x_data**2+y_data**2)<1)/N_monte

#解析解(円部分の面積/正方形の面積)
true_rate=(np.pi*1*1/4)/(1*1)

print('求めた値：',sampling_rate)
print('正解値：',true_rate)
```

```python
#forループで乱数の数を変えて計算
N_monte_list=np.arange(1000,500000+1,1000)

#結果を格納するリストを作成
sampling_rate_list=[]

for N_monte in N_monte_list:
    x_data=np.random.rand(N_monte)
    y_data=np.random.rand(N_monte)
    
    sampling_rate=np.sum((x_data**2+y_data**2)<1)/N_monte
    
    sampling_rate_list.append(sampling_rate)

plt.plot(N_monte_list,sampling_rate_list)
plt.xlabel('N_monte')
plt.savefig('./figure/fig_montecalro.jpg')
plt.show()
```
